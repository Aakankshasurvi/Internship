```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', class_='wikitable')
rows = table.find_all('tr')[1:]  # Exclude header row

for row in rows:
    cells = row.find_all('td')
    if len(cells) < 5:
        print("Incomplete row:", row)
        continue
    
    rank = cells[0].text.strip()
    name = cells[1].text.strip()
    artist = cells[2].text.strip()
    upload_date = cells[3].text.strip()
    views = cells[4].text.strip()
    
    print("Rank:", rank)
    print("Name:", name)
    print("Artist:", artist)
    print("Upload Date:", upload_date)
    print("Views:", views)
    print()
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the BCCI homepage
url = "https://www.bcci.tv/"

# Send a GET request to the homepage
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find the link to the international fixtures page
fixtures_link = soup.find('a', string="Fixtures")

# Check if the link is found
if fixtures_link:
    # Extract the URL of the fixtures page
    fixtures_url = fixtures_link('href')
    
    # Construct the full URL of the fixtures page
    fixtures_url = f'https://www.bcci.tv/fixtures?platform=international&type=men'
    
    # Send a GET request to the fixtures page
    fixtures_response = requests.get(fixtures_url)
    
    # Parse the HTML content of the fixtures page
    fixtures_soup = BeautifulSoup(fixtures_response.text, 'html.parser')
    
    # Find all the fixtures
    fixtures = fixtures_soup.find_all('div', class_='fixture__info')
    
    # Iterate through each fixture and extract the details
    for fixture in fixtures:
        series = fixture.find('span', class_='u-unskewed-text').text.strip()
        place = fixture.find('p', class_='fixture__additional-info').text.strip()
        date = fixture.find('span', class_='fixture__date').text.strip()
        time = fixture.find('span', class_='fixture__time').text.strip()
        
        print("Series:", series)
        print("Place:", place)
        print("Date:", date)
        print("Time:", time)
        print("="*50)
else:
    print("Link to International Fixtures not found on the homepage.")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the GitHub trending page
url = "https://github.com/trending"

# Send a GET request to the website
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the list of trending repositories
repositories = soup.find_all("article", class_="Box-row")

# Iterate over each repository to extract details
for repo in repositories:
    # Repository title
    title_elem = repo.find("h1", class_="h3 lh-condensed")
    title = title_elem.text.strip() if title_elem else "No title"

    # Repository description
    description_elem = repo.find("p", class_="col-9 color-text-secondary my-1 pr-4")
    description = description_elem.text.strip() if description_elem else "No description"

    # Contributors count
    contributors_elem = repo.find("a", class_="Link--muted d-inline-block mr-3")
    contributors_count = contributors_elem.text.strip() if contributors_elem else "No contributors"

    # Language used
    language_elem = repo.find("span", class_="d-inline-block ml-0 mr-3")
    language = language_elem.text.strip() if language_elem else "No language"

    # Print the details
    print("Repository title:", title)
    print("Repository description:", description)
    print("Contributors count:", contributors_count)
    print("Language used:", language)
    print("-----------------------------------")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the UCI Machine Learning Repository
base_url = 'https://archive.ics.uci.edu/ml/'

# Send a GET request to the UCI Machine Learning Repository main page
response = requests.get(base_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
   
    # Find all paragraphs (<p>) on the page
    paragraphs = soup.find_all('p')
   
    # Iterate through paragraphs to find dataset-related information
    for paragraph in paragraphs:
        # Check if the paragraph contains dataset details
        if 'Data Set' in paragraph.text or 'Dataset' in paragraph.text:
            # Extract dataset information
            dataset_info = paragraph.text.strip()
           
            # Print the extracted dataset information
            print(dataset_info)
            print("------------------------")
else:
    print(f"Failed to retrieve UCI Machine Learning Repository page. Status Code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"

# Send a GET request to the website
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the highest-selling novels data
table = soup.find("table", class_="in-article sortable")

# Initialize lists to store the scraped data
book_names = []
author_names = []
volumes_sold = []
publishers = []
genres = []

# Extract data from each row of the table
for row in table.find_all("tr")[1:]:
    # Extract data from each column of the row
    columns = row.find_all("td")
    
    # Check if the row contains data for all columns
    if len(columns) >= 5:
        # Append data to respective lists
        book_names.append(columns[0].text.strip())
        author_names.append(columns[1].text.strip())
        volumes_sold.append(columns[2].text.strip())
        publishers.append(columns[3].text.strip())
        genres.append(columns[4].text.strip())

# Print the scraped data
for book_name, author_name, volume_sold, publisher, genre in zip(book_names, author_names, volumes_sold, publishers, genres):
    print("Book name:", book_name)
    print("Author name:", author_name)
    print("Volumes sold:", volume_sold)
    print("Publisher:", publisher)
    print("Genre:", genre)
    print("-----------------------------")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the IMDb page
url = "https://www.imdb.com/list/ls512407256/"

# Send a GET request to the website
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the list of TV series
series_list = soup.find_all("div", class_="lister-item-content")

# Initialize lists to store the scraped data
names = []
year_spans = []
genres = []
runtimes = []
ratings = []
votes = []

# Extract data from each TV series
for series in series_list:
    # Name
    name_elem = series.find("h3", class_="lister-item-header").find("a")
    name = name_elem.text.strip() if name_elem else "N/A"
    names.append(name)

    # Year span
    year_span_elem = series.find("span", class_="lister-item-year")
    year_span = year_span_elem.text.strip() if year_span_elem else "N/A"
    year_spans.append(year_span)

    # Genre
    genre_elem = series.find("span", class_="genre")
    genre = genre_elem.text.strip() if genre_elem else "N/A"
    genres.append(genre)

    # Run time
    runtime_elem = series.find("span", class_="runtime")
    runtime = runtime_elem.text.strip() if runtime_elem else "N/A"
    runtimes.append(runtime)

    # Ratings
    rating_elem = series.find("div", class_="ipl-rating-star")
    rating = rating_elem.text.strip() if rating_elem else "N/A"
    ratings.append(rating)

    # Votes
    vote_elem = series.find("span", attrs={"name": "nv"})
    vote = vote_elem.text.strip() if vote_elem else "N/A"
    votes.append(vote)

# Print the scraped data
for name, year_span, genre, runtime, rating, vote in zip(names, year_spans, genres, runtimes, ratings, votes):
    print("Name:", name)
    print("Year span:", year_span)
    print("Genre:", genre)
    print("Run time:", runtime)
    print("Ratings:", rating)
    print("Votes:", vote)
    print("-----------------------------")
```

```python

import requests
from bs4 import BeautifulSoup

url = 'https://www.imdb.com/list/ls512407256/'

# Specify user-agent header to mimic a legitimate browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Send a GET request to the IMDb URL with headers
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all TV series items in the list
    series_items = soup.find_all('div', class_='lister-item-content')

    for series in series_items:
        try:
            # Extract name of the series
            name = series.find('h3', class_='lister-item-header').a.text.strip()

            # Extract year span
            year_span = series.find('span', class_='lister-item-year').text.strip()

            # Extract genre
            genre = series.find('span', class_='genre').text.strip()

            # Extract runtime
            runtime = series.find('span', class_='runtime').text.strip()

            # Extract rating
            rating = series.find('span', class_='ipl-rating-star__rating').text.strip()

            # Extract votes
            votes = series.find('span', attrs={'name': 'nv'}).text.strip()

            # Print the extracted details
            print(f"Name: {name}")
            print(f"Year Span: {year_span}")
            print(f"Genre: {genre}")
            print(f"Run Time: {runtime}")
            print(f"Rating: {rating}")
            print(f"Votes: {votes}")
            print("------------------------")

        except Exception as e:
            print(f"Error occurred while extracting data: {e}")

else:
    print(f"Failed to retrieve IMDb page. Status Code: {response.status_code}")
```
